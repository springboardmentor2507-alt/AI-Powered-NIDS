# MAPPING ATTACK FEATURES

```python
dos = ['back','land','neptune','pod','smurf','teardrop']
probe = ['satan','ipsweep','nmap','portsweep']
r2l = ['ftp_write','guess_passwd','imap','multihop','phf','spy','warezclient','warezmaster']
u2r = ['buffer_overflow','loadmodule','perl','rootkit']

def map_attack(x):
    if x == 'normal':
        return 'normal'
    elif x in dos:
        return 'dos'
    elif x in probe:
        return 'probe'
    elif x in r2l:
        return 'r2l'
    elif x in u2r:
        return 'u2r'
    else:
        return 'other'   # For safety

train_df['attack_category'] = train_df['target'].apply(map_attack)
```

```python
ct = pd.crosstab(train_df['attack_category'], train_df['service'])

top5_services = train_df['service'].value_counts().head(5).index
ct = ct[top5_services] 

ct_percentage = ct.div(ct.sum(axis=1), axis=0) * 100

plt.figure(figsize=(18, 8))

bottom = None

for column in ct_percentage.columns:
    if bottom is None:
        plt.barh(ct_percentage.index, ct_percentage[column], label=column)
        bottom = ct_percentage[column].copy()
    else:
        plt.barh(ct_percentage.index, ct_percentage[column], left=bottom, label=column)
        bottom += ct_percentage[column]

for i, attack in enumerate(ct_percentage.index):
    for j, column in enumerate(ct_percentage.columns):
        value = ct_percentage.loc[attack, column]
        if value > 2:  # show only if significant
            plt.text(
                ct_percentage.iloc[i, :j+1].sum() - value/2,
                i,
                f"{value:.1f}%",
                ha='center',
                va='center',
                fontsize=10,
                color='black'
            )

plt.title("Attack Distribution by Top 5 Services", fontsize=18, weight='bold')
plt.xlabel("Percentage (%)", fontsize=12)
plt.ylabel("Attack Category", fontsize=12)
plt.legend(title="Service Type", bbox_to_anchor=(1.02, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

# CONVERTING CATEGORICAL FEATURES USING ENCODING TECHNIQUE

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

train_df['attack_category_encoded'] = le.fit_transform(train_df['attack_category'])
print(train_df[['attack_category', 'attack_category_encoded']].head())

print("\nEncoding Mapping:")
mapping = dict(zip(le.classes_, le.transform(le.classes_)))
for label, code in mapping.items():
    print(f"{label} -> {code}")
```

# Normalize the numerical features.  

```python
from sklearn.preprocessing import MinMaxScaler

num_cols = train_df.select_dtypes(include=['int64', 'float64']).columns
num_cols = [c for c in num_cols if c not in ['protocol_type','service','flag','attack_category']]

scaler = MinMaxScaler()
train_df[num_cols] = scaler.fit_transform(train_df[num_cols])
```

# SPLITING TRAIN AND TEST SET

```python
from sklearn.model_selection import train_test_split

# Split the dataset
train_df, test_df = train_test_split(
    train_df,
    test_size=0.2,                # 20% test set            
    stratify=train_df['attack_category'] 
)

print("Training shape:", train_df.shape)
print("Testing shape :", test_df.shape)
```
